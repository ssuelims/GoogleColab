```python
# funçao loc() e iloc()
# todas as linhas : e 0 - indice da primeira coluna
```

```python
x_base_age = df.iloc[:,0].values
```

```python
print(f" age: {x_base_age}")
```

```python
y_base_Years_of_Experience = df.iloc[:,4].values
```

```python
print(f" Years_of_Experience: {y_base_Years_of_Experience}")
```

# correlaçao de Variaveis
-Implicaçao direta,constante

```python
import numpy as np
```

```python
np.corrcoef(x_base_age,y_base_Years_of_Experience)
```

# Modelo Preditivo

```python
from sklearn.linear_model import LinearRegression
```

```python
# criar um objeto
reg_age = LinearRegression()
```

```python
x_base_age = x_base_age.reshape(-1,1)
```

```python
reg_age.fit(x_base_age,y_base_Years_of_Experience)
```

# Previsao

```python
print(f"Years_of_Experience {reg_age.predict([[20]])}")
```

```python
print(f"Years_of_Experience {reg_age.predict([[60]])}")
```
